```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters for the hyperbola
a = 3  # Semi-major axis
b = 2  # Semi-minor axis

# Define x values for the hyperbola branches
x = np.linspace(-5, 5, 500)

# Calculate y values for both branches of the hyperbola
y_positive = b * np.sqrt((x**2 / a**2) - 1)
y_negative = -b * np.sqrt((x**2 / a**2) - 1)

# Mask values where the hyperbola is not defined
y_positive = np.where(x**2 > a**2, y_positive, np.nan)
y_negative = np.where(x**2 > a**2, y_negative, np.nan)

# Plot the hyperbola
plt.figure(figsize=(8, 6))
plt.plot(x, y_positive, label="Top Branch", color="blue")
plt.plot(x, y_negative, label="Bottom Branch", color="red")

# Add axes
plt.axhline(0, color="black", linewidth=0.5)
plt.axvline(0, color="black", linewidth=0.5)

# Label the graph
plt.title("Hyperbola: $\\frac{x^2}{a^2} - \\frac{y^2}{b^2} = 1$")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters for the hyperbola
a = 3  # Semi-major axis
b = 2  # Semi-minor axis

# Define x values for the hyperbola branches, ensuring x^2 >= a^2
x_positive = np.linspace(a, 5, 500)  # Right branch
x_negative = np.linspace(-5, -a, 500)  # Left branch

# Calculate y values for both branches of the hyperbola
y_positive = b * np.sqrt((x_positive**2 / a**2) - 1)
y_negative = -b * np.sqrt((x_positive**2 / a**2) - 1)

# Plot the hyperbola
plt.figure(figsize=(8, 6))
plt.plot(x_positive, y_positive, label="Top Right Branch", color="blue")
plt.plot(x_positive, y_negative, label="Bottom Right Branch", color="red")
plt.plot(x_negative, y_positive, label="Top Left Branch", color="green")
plt.plot(x_negative, y_negative, label="Bottom Left Branch", color="orange")

# Add axes
plt.axhline(0, color="black", linewidth=0.5)
plt.axvline(0, color="black", linewidth=0.5)

# Label the graph
plt.title("Hyperbola: $\\frac{x^2}{a^2} - \\frac{y^2}{b^2} = 1$")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()
```
